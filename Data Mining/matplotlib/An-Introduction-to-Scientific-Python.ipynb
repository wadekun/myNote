# matplotlib
matplotlib是一个python绘图库。它包含了大量的绘图工具，我们可以使用这些工具创建各种图形，包括散点图，正弦曲线图，甚至三维图。

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(20)
y = x**2

plt.plot(x, y)
```

```python
# 画一个 [0, 2pi] 上的正弦曲线图
x = np.linspace(0, 2 * np.pi, 50)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

## 在一张图像上画两个数据集

```python
x = np.linspace(0, 2 * np.pi, 50)
plt.plot(x, np.sin(x), x, np.sin(2 * x))
plt.show()
```

# 自定义图形的外观

当多个数据集在同一个图像上时，通过改变线条的外观来区分不同的数据集变得非常必要。

```python
x = np.linspace(0, 2 * np.pi, 50)
plt.plot(x, np.sin(x), 'r-o', 
        x, np.sin(2 * x), 'g--')
plt.show()
```

`r-o`，`g--`，其中`r`、`g`代表线条的颜色。`-o`表示实心点标记的实线，`--`代表虚线。

## 使用子图
使用子图可以在一个窗口中绘制多张图。

```python
# 使用子图
x = np.linspace(0, 2 * np.pi, 50)
plt.subplot(2, 1, 1) # 行，列，活跃区
plt.plot(x, np.sin(x), 'r')
plt.subplot(2, 1, 2) #行，列，活跃区
plt.plot(x, np.cos(x), 'g')
plt.show()
```

使用子图只需要一个额外的步骤，就可以像前面的例子一样绘制数据集。即在调用 `plot()` 函数之前需要先调用 `subplot()` 函数。该函数的第一个参数代表子图的总行数，第二个参数代表子图的总列数，第三个参数代表活跃区域。

活跃区域代表当前子图所在绘图区域，绘图区域是按从左至右，从上至下的顺序编号。例如在 `4×4` 的方格上，活跃区域 `6` 在方格上的坐标为 `(2, 2)`。

# 简单的散点图
散点图就是一堆离散点的集合。

```python
# 简单散点图
x = np.linspace(0, 2 * np.pi, 50)
y = np.sin(x)
plt.scatter(x, y)
plt.show()
```

# 彩色映射散点图

```python
#彩色散点图
x = np.random.rand(1000)
y = np.random.rand(1000)
size = np.random.rand(1000) * 50
color = np.random.rand(1000)
plt.scatter(x, y, size, color)
plt.colorbar()
plt.show()
```

# 直方图

```python
# 直方图
x = np.random.randn(1000) # 数据
plt.hist(x, 50) # 将数据分为50分，直方图展示每份的数据量
plt.show()
```

# 标题标签和图例

```python
x = np.linspace(0, 2 * np.pi, 50)
plt.plot(x, np.sin(x), 'r-o', label='sin(x)')
plt.plot(x, np.cos(x), 'g--', label='cos(x)')
plt.legend()
plt.xlabel('Rads') # 给x轴添加标签
plt.ylabel('Amplitude') # 给y轴添加标签
plt.title('Sin and Cos Waves')
plt.show()
```
